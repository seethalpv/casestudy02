# 1. Find out how many males and females participated in the test.

# 2. What do you think about the students parental level of education?

```python
data[['gender','parental level of education','math score','writing score','reading score']].groupby(['gender','parental level of education']).agg('median')
```

```python
Insights:
Among students with the same parental level of education, 
female students have higher average reading and writing scores than their male counterparts. 
male students have higher average math score.
```

# 3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
data.groupby(['gender']).agg(['mean'])
```

```python
Insights:
    Female scores most on reading and writing
    male in maths
```

```python
data.groupby(['test preparation course']).agg(['mean'])
```

```python
Insights:
    According to test preparation course,
    Test preparation course completed students having higher scores.
```
